# U.S. Medical Insurance Costs

```python
import csv


insurance = []
with open('insurance.csv', newline = '') as insurance_data:
    data = csv.DictReader(insurance_data)
    for row in data:
        insurance.append(row)

def variable_maker(variable, data):
    lst = []
    for row in data:
        lst.append(row[variable])
    return lst
age = variable_maker('age', insurance)
sex = variable_maker('sex', insurance)
bmi = variable_maker('bmi', insurance)
children = variable_maker('children', insurance)
smoker = variable_maker('smoker', insurance)
region = variable_maker('region', insurance)
charges = variable_maker('charges', insurance)


class Persons():
    
    def __init__(self, age, sex, bmi, children, smoker, region, charges):
        self.age = age
        self.sex = sex
        self.bmi = bmi
        self.children = children
        self.smoker = smoker
        self.region = region
        self.charges = charges
    
    def average_cost_smoker_vs_non_smoker(self):
        
        total_cost_smoker = 0
        total_cost_non_smoker = 0
        counter_smoker = 0
        counter_non_smoker = 0
        data = zip(self.smoker, self.charges)
        #determines if someone is a smoker and adds the cost to the totals
        for i in data:
            if i[0] == 'yes':
                total_cost_smoker += float(i[1])
                counter_smoker += 1
            if i[0] == 'no':
                total_cost_non_smoker += float(i[1])
                counter_non_smoker += 1
        #determines the average cost of smokers and non-smokers        
        avg_cost_smoker = total_cost_smoker / counter_smoker
        avg_cost_non_smoker = total_cost_non_smoker / counter_non_smoker
        difference = round(avg_cost_smoker - avg_cost_non_smoker, 2)
        return difference
    
    def area_of_smokers(self):
        people_per_area = {}
        smokers_per_area = {}
        data = zip(self.smoker, self.region)
        data1 = self.region
        #people per erea
        for area in data1:
            if area in people_per_area:
                people_per_area[area] += 1
            else:
                people_per_area[area] = 1
                    
        #smokers per erea
        for i in data:
            if i[0] == 'yes':
                if i[1] in smokers_per_area:
                    smokers_per_area[i[1]] += 1
                else:
                    smokers_per_area[i[1]] = 1
            if i[0] == 'no':
                pass
            
        #percentage of smokers per area
        perc_smokers_area = {}
        for key, value in people_per_area.items():
            perc_smokers_area[key] = round((smokers_per_area[key]/value)*100, 2)
            
        return perc_smokers_area

    def smokers_with_children(self):
        no_children = 0
        no_children_smokers = 0 
        one_child = 0
        one_child_smokers = 0
        more_than_one_child = 0
        more_than_one_child_smokers = 0
        data = zip(self.children, self.smoker)
        data1 = zip(self.children, self.smoker)
        #determines how many people there are with 0, 1 are more children
        for person in data:
            if person[0] == '0':
                no_children += 1
            if person[0] == '1':
                one_child += 1
            else:
                more_than_one_child += 1 
        #determines how many people with certain amount of children smoke
        for person in data1:
            if person[0] == '0' and person[1] == 'yes':
                no_children_smokers += 1
            if person[0] == '1' and person[1] == 'yes':
                one_child_smokers += 1
            if person[1] == 'no':
                pass
            else:
                more_than_one_child_smokers += 1
        
        percentage_no_children = round(no_children_smokers / no_children * 100, 2)
        percentage_one_child = round(one_child_smokers / one_child * 100, 2)
        percentage_more_than_one_child = round(more_than_one_child_smokers / more_than_one_child * 100, 2)
        
        return  percentage_no_children, percentage_one_child, percentage_more_than_one_child
        
        
        
        

insurance_costs = Persons(age, sex, bmi, children, smoker, region, charges)

cost_of_smoking = insurance_costs.average_cost_smoker_vs_non_smoker()
print("The average extra costs in health insurance by smoking is " + str(cost_of_smoking) + " dollars.")

regions_with_most_smokers = insurance_costs.area_of_smokers()
print("This is the percentage of people that smoke, according to the region they live in: " + str(regions_with_most_smokers))

smokers_with_children = insurance_costs.smokers_with_children()
print("The percentage of people that smoke with 0, 1 or more children respectively is: " + str(smokers_with_children))

       
```
